```
# свертка цветного изображения
```

```
def filter(x):
  return np.mean(x)
win_size= (16,16)
new_image=[]
for c in range(3):
  out_image=np.zeros((im.shape[0]-win_size[0]+1,  im.shape[1]-win_size[1]+1),dtype=np.float32)
  for i in range(out_image.shape[0]):
    for j in range(out_image.shape[1]):
      out_image[i,j]=filter(im[i:i+win_size[0],j:j+win_size[1], c])
     
  print(out_image.shape)
  new_image.append(out_image)
```

```
new_image=np.stack(new_image, axis=2)
```

```
plt.imshow(new_image)
```
